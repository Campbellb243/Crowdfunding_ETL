### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
columns_list = crowdfunding_info_df.columns.tolist()
print(columns_list)
```

```python
# Assign the category and subcategory values to category and subcategory columns.

# Splitting the values of 'category & sub-category' column based on '&'
category_subcategory = crowdfunding_info_df['category & sub-category'].str.split('&', expand=True)

# Assigning the first part to 'category' column and the second part to 'subcategory' column
crowdfunding_info_df['category'] = category_subcategory[0].str.strip()
crowdfunding_info_df['subcategory'] = category_subcategory[1].str.strip()

# Drop the original 'category & sub-category' column if needed
# crowdfunding_info_df.drop(columns=['category & sub-category'], inplace=True)

# Displaying the first few rows of the DataFrame to verify the changes
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.

# Splitting the values of 'category & sub-category' column based on '&'
category_subcategory = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)

# Extracting unique categories and subcategories
unique_categories = category_subcategory[0].str.strip().unique().tolist()
unique_subcategories = category_subcategory[1].str.strip().unique().tolist()

print("Unique categories:")
print(unique_categories)

print("\nUnique subcategories:")
print(unique_subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.

# Assuming you have already obtained the unique categories and subcategories lists

# Print the number of distinct values in the categories and subcategories lists
print("Number of distinct categories:", len(unique_categories))
print("Number of distinct subcategories:", len(unique_subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
# Create numpy arrays for category_ids and subcategory_ids
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

# Print the created numpy arrays
print("Category IDs:")
print(category_ids)

print("\nSubcategory IDs:")
print(subcategory_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.

# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
# Add "cat" to each category ID using list comprehension
cat_ids = ["cat" + str(cat_id) for cat_id in category_ids]

# Add "subcat" to each subcategory ID using list comprehension
scat_ids = ["subcat" + str(scat_id) for scat_id in subcategory_ids]

# Print the modified lists
print("Category IDs with 'cat':")
print(cat_ids)

print("\nSubcategory IDs with 'subcat':")
print(scat_ids)
```

```python
import pandas as pd

# Create a DataFrame for categories
category_df = pd.DataFrame({
    'category_id': cat_ids,
    'category_name': unique_categories
})

# Create a DataFrame for subcategories
subcategory_df = pd.DataFrame({
    'subcategory_id': scat_ids,
    'subcategory_name': unique_subcategories
})

# Print the created DataFrames
print("Category DataFrame:")
display(category_df)

print("\nSubcategory DataFrame:")
subcategory_df

```
